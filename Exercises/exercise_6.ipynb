```python
import os
import numpy as np

from si.io.csv_file import read_csv
from si.model_selection.split import stratified_train_test_split

IRIS_PATH = os.path.join('..', 'datasets', 'iris', 'iris.csv')
```

```python
# Load dataset
# Read CSV into Dataset (features + label)
ds = read_csv(filename=IRIS_PATH, features=True, label=True)

# Perform stratified split (preserve class proportions)
train, test = stratified_train_test_split(ds, test_size=0.3, random_state=0)

# Helper to count class occurrences
def counts(y):
    unique, c = np.unique(y, return_counts=True)
    return dict(zip(unique, c))

# Show total vs. train/test class distribution
print('Total counts:', counts(ds.y))
print('Train counts:', counts(train.y))
print('Test counts:', counts(test.y))
print('Train shape:', train.X.shape, 'Test shape:', test.X.shape)
```
