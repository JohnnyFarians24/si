```python
import os
import numpy as np

from si.io.csv_file import read_csv
from si.decomposition.pca import PCA

IRIS_PATH = os.path.join('..', 'datasets', 'iris', 'iris.csv')
```

```python
# Load iris dataset
# Read CSV into Dataset (features + label)
ds = read_csv(filename=IRIS_PATH, features=True, label=True)
print('Original shape:', ds.X.shape)  # (n_samples, n_features)
print('First 3 rows of X:', ds.X[:3])
```

```python
# Fit PCA and inspect results
pca = PCA(n_components=2)          # keep first 2 principal components
pca.fit(ds)                        # compute mean, components, explained variance
ds_red = pca.transform(ds)         # project original data to 2D space

# Display shapes and a preview
print('Reduced shape:', ds_red.X.shape)  # (n_samples, 2)
print('Reduced first 3 rows:', ds_red.X[:3])
print('Components shape:', pca.components.shape)  # (2, n_features)
print('Explained variance:', pca.explained_variance)  # variance fraction per PC
print('Explained variance sum:', float(np.sum(pca.explained_variance)))
```
