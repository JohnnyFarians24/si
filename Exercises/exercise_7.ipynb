```python
import os
import numpy as np

from si.io.csv_file import read_csv
from si.models.knn_regressor import KNNRegressor
from si.model_selection.split import train_test_split

CPU_PATH = os.path.join('..', 'datasets', 'cpu', 'cpu.csv')
```

```python
# Load CPU dataset
# Read CSV into Dataset (features + label)
ds = read_csv(filename=CPU_PATH, features=True, label=True)
print('Dataset shape:', ds.X.shape)  # (n_samples, n_features)
print('First 3 rows:', ds.X[:3])
```

```python
# Split dataset (random train/test)
train, test = train_test_split(ds, test_size=0.2, random_state=0)

# Fit k-NN regressor (k=5)
knn = KNNRegressor(k=5)
knn.fit(train)

# Predict on test set
preds = knn.predict(test)

# Evaluate prediction error (RMSE)
rmse_val = knn.score(test)
print('RMSE:', rmse_val)
```
