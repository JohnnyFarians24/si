```python
import os

from si.io.csv_file import read_csv
from si.model_selection.split import train_test_split
from si.models.random_forest_classifier import RandomForestClassifier

IRIS_PATH = os.path.join('..', 'datasets', 'iris', 'iris.csv')
```

```python
iris_ds = read_csv(filename=IRIS_PATH, features=True, label=True)
print('Original shape:', iris_ds.X.shape)

# Split into train/test 
train, test = train_test_split(iris_ds, test_size=0.3, random_state=0)

# Initialize random forest (50 trees, depth 5)
rf = RandomForestClassifier(n_estimators=50, max_depth=5, seed=0)
rf.fit(train)  # train ensemble

# Evaluate accuracy on test set
score = rf.score(test)
print('RandomForest test accuracy:', score)
```
