```python
import os
import numpy as np

from si.io.csv_file import read_csv
from si.feature_selection.select_percentile import SelectPercentile
from si.statistics.f_classification import f_classification

IRIS_PATH = os.path.join('..', 'datasets', 'iris', 'iris.csv')
```

```python
ds = read_csv(filename=IRIS_PATH, features=True, label=True)
print('Original shape:', ds.X.shape)
print('Features:', ds.features)
```

```python
# Select top 40% features by ANOVA F-score
selector = SelectPercentile(score_func=f_classification, percentile=40)
selector.fit(ds)
ds_sel = selector.transform(ds)
print('Selected shape:', ds_sel.X.shape)
print('Selected features:', ds_sel.features)
print('F scores:', selector.F)
print('p-values:', selector.p)
```
